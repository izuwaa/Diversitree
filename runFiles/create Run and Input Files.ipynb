```python
cwd = os.getcwd() + '/Train'
print(cwd)

for percent in percentages:
    
    for problem in trainSet:
        
        for j in numNearOptimal:
            
            fName = percent + str(j) + problem
            testFileName = percent + str(j) + problem + '.txt'
            fullFName = cwd  +'/' + fName
            
            print('writing ',fName, '\n')
            with open(fullFName,'w') as f:
                f.write('#!/bin/bash')
                f.write('\n')
                f.write('#')
                f.write('\n')
                f.write('#PBS -N ' + percent + fName)
                f.write('\n')
                f.write('#PBS -M iahanor@vols.utk.edu')
                f.write('\n')
                f.write('#PBS -m abe')
                f.write('\n')
                f.write('#PBS -A ACF-UTK0011')
                f.write('\n')
                f.write('#PBS -o /lustre/haven/user/iahanor/branchandbound/Train/' +  percent + '/' + fName + 'O.txt')
                f.write('\n')
                f.write('#PBS -e /lustre/haven/user/iahanor/branchandbound/Train/' + percent + '/' + fName + 'E.txt')
                f.write('\n')
                f.write('#PBS -l walltime=24:00:00')
                f.write('\n')
                f.write('#PBS -l nodes=1:ppn=4')
                f.write('\n')
                f.write('#PBS -l partition=gpu')
                f.write('\n')
                f.write('#PBS -l feature=cascadelake_gpu')
                f.write('\n')
                f.write('\n')
                f.write('cd /lustre/haven/user/iahanor/branchandbound/Train/' + percent)
                f.write('\n')
                f.write('mkdir ' + fName + 'F')
                f.write('\n')
                f.write('cd ' + fName + 'F')
                f.write('\n')
                f.write('\n')
                f.write('git clone https://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@github.com/izuwaa/Diversitree.git')
                f.write('\n')
                f.write('cd Diversitree')
                f.write('\n')
                f.write('module load gcc/6.3.0')
                f.write('\n')
                f.write('module load cmake')
                f.write('\n')
                f.write('\n')
                f.write('cmake -Bbuild -H. -DSCIP_DIR=/lustre/haven/user/iahanor/branchandbound/scipoptsuite-7.0.3/installDir/lib/cmake/scip -DGMP_INCLUDE_DIR=/sw/cs400_centos7.3_acfsoftware/gmp/5.1.1/centos7.3_intel17.2.174/include')
                f.write('\n')
                f.write('cmake --build build')
                f.write('\n')
                f.write('echo "module loaded"')
                f.write('\n')
                f.write('\n')
                f.write('cd /lustre/haven/user/iahanor/branchandbound/Train/' + percent + '/' + fName + 'F' + '/' + 'Diversitree')
                f.write('\n')
                f.write('/lustre/haven/user/iahanor/branchandbound/Train/' + percent + '/' + fName + 'F' + '/' + 'Diversitree/build/diversitree ' + testFileName + ' problemOptions , solverParams.txt solverOptions :')
                f.write('\n')
                f.write('echo "module I am done"')
```
